```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 입력
sizes = [10, 100, 1000]

# For loop 실행 시간 (평균값 사용)
for_loop_means = [
    np.mean([0.0010113716125488281, 0.0, 0.0010156631469726562, 0.0009882450103759766, 0.0009958744049072266]),  # 10x10
    np.mean([0.05598115921020508, 0.057127952575683594, 0.04649496078491211, 0.04829597473144531, 0.0468144416809082]),  # 100x100
    np.mean([6.122434616088867, 6.175225734710693, 6.0545666217803955, 6.074216365814209, 5.899410724639893])  # 1000x1000
]

# Vectorized 실행 시간 (평균값 사용) - 첫 값은 0으로 설정
vectorized_means = [
    0.0,  # 10x10
    np.mean([0.00590205192565918, 0.0070993900299072266, 0.007688045501708984, 0.006602764129638672, 0.0077381134033203125]),  # 100x100
    np.mean([77.12131547927856, 75.42761063575745, 87.55694627761841, 87.24637269973755, 98.71136713027954])  # 1000x1000
]

# Plot 설정 (합쳐서 그리기)
plt.figure(figsize=(10, 6))

# For loop 실행 시간 그래프
plt.plot(sizes, for_loop_means, label='For Loop', marker='o', color='blue')

# Vectorized 실행 시간 그래프
plt.plot(sizes, vectorized_means, label='Vectorized', marker='x', color='red')

# Plot 속성 설정
plt.xlabel('Matrix Size')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time: For Loop vs Vectorized')
plt.legend()
plt.ylim(0, 100)  # y축을 0 ~ 100으로 설정
plt.grid(True)

# Plot 표시
plt.show()
```

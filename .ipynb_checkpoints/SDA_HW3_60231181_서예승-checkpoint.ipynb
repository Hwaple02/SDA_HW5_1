```python
#Q1
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")

plt.figure(figsize=(12, 6))
df.boxplot(column='target', by='base_hour', grid=False)
plt.title('Traffic Volume Distribution by Hour')
plt.suptitle('')
plt.xlabel('Base Hour')
plt.ylabel('Traffic Volume (Target)')
plt.xticks(rotation=90)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False 

df = pd.read_csv("train.csv")

day_order = ['월', '화', '수', '목', '금', '토', '일']
df['day_of_week'] = pd.Categorical(df['day_of_week'], categories=day_order, ordered=True)

plt.figure(figsize=(12, 6))
df.boxplot(column='target', by='day_of_week', grid=False)  
plt.title('요일별 교통량 분포', fontsize=14)
plt.suptitle('')  
plt.xlabel('요일', fontsize=12)
plt.ylabel('교통량 (Target)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)  
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  
plt.show()
```

```python
#Q3
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")

df['year_month'] = df['base_date'].astype(str).str[:6] 

plt.figure(figsize=(14, 8))
df.boxplot(column='target', by='year_month', grid=False)
plt.title('Traffic Volume Distribution by Year and Month')
plt.suptitle('') 
plt.xlabel('Year-Month')
plt.ylabel('Traffic Volume (Target)')
plt.xticks(rotation=90) 
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Q7
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")

heatmap_data = df.pivot_table(
    values='target',  
    index='lane_count',  
    columns='base_hour',  
    aggfunc='mean' 
)

plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="coolwarm", linewidths=0.5)
plt.title('Heatmap: Base Hour vs. Lane Count (Traffic Volume)', fontsize=16)
plt.xlabel('Base Hour (시간)')
plt.ylabel('Lane Count (차선 수)')
plt.show()
```

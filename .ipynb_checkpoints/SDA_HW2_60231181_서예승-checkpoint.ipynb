```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import time

# 1. Generate sample data from two different distributions and plot them
np.random.seed(42)

# Generating 100,000 samples from Normal Distribution (mean=0, std=1)
sample_1 = np.random.normal(0, 1, 100000)

# Generating 100,000 samples from Uniform Distribution (low=-3, high=3)
sample_2 = np.random.uniform(-3, 3, 100000)

# Plotting the distributions
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(sample_1, bins=50, color='skyblue', alpha=0.7)
plt.title('Normal Distribution (mean=0, std=1)')

plt.subplot(1, 2, 2)
plt.hist(sample_2, bins=50, color='salmon', alpha=0.7)
plt.title('Uniform Distribution (low=-3, high=3)')

plt.tight_layout()
plt.show()

# 2. Calculating statistics with and without vectorization

def calculate_statistics(samples):
    mean = np.mean(samples)
    median = np.median(samples)
    variance = np.var(samples)
    mad = np.mean(np.abs(samples - mean))
    skewness = stats.skew(samples)
    kurtosis = stats.kurtosis(samples)
    iqr = np.percentile(samples, 75) - np.percentile(samples, 25)
    return mean, median, variance, mad, skewness, kurtosis, iqr

# Without vectorization (manual implementation)
def calculate_statistics_non_vectorized(samples):
    n = len(samples)
    mean = sum(samples) / n
    sorted_samples = sorted(samples)
    median = sorted_samples[n // 2] if n % 2 != 0 else (sorted_samples[n // 2 - 1] + sorted_samples[n // 2]) / 2
    variance = sum((x - mean) ** 2 for x in samples) / n
    mad = sum(abs(x - mean) for x in samples) / n
    skewness = stats.skew(samples)
    kurtosis = stats.kurtosis(samples)
    q1, q3 = np.percentile(samples, [25, 75])
    iqr = q3 - q1
    return mean, median, variance, mad, skewness, kurtosis, iqr

# Time comparison for vectorized vs non-vectorized
start_time = time.time()
statistics_vectorized = calculate_statistics(sample_1)
vectorized_time = time.time() - start_time

start_time = time.time()
statistics_non_vectorized = calculate_statistics_non_vectorized(sample_1)
non_vectorized_time = time.time() - start_time

# Printing the results
print("Vectorized - Time taken: {:.5f} seconds".format(vectorized_time))
print("Non-vectorized - Time taken: {:.5f} seconds".format(non_vectorized_time))
print("\nStatistics (Vectorized):")
print(f"Mean: {statistics_vectorized[0]:.5f}, Median: {statistics_vectorized[1]:.5f}, Variance: {statistics_vectorized[2]:.5f}, MAD: {statistics_vectorized[3]:.5f}, Skewness: {statistics_vectorized[4]:.5f}, Kurtosis: {statistics_vectorized[5]:.5f}, IQR: {statistics_vectorized[6]:.5f}")

print("\nStatistics (Non-vectorized):")
print(f"Mean: {statistics_non_vectorized[0]:.5f}, Median: {statistics_non_vectorized[1]:.5f}, Variance: {statistics_non_vectorized[2]:.5f}, MAD: {statistics_non_vectorized[3]:.5f}, Skewness: {statistics_non_vectorized[4]:.5f}, Kurtosis: {statistics_non_vectorized[5]:.5f}, IQR: {statistics_non_vectorized[6]:.5f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib as mpl
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# Set font to NanumGothic
mpl.rcParams['font.family'] = 'NanumGothic'
mpl.rcParams['axes.unicode_minus'] = False  # Prevent issues with minus sign

# Load the data
data = pd.read_csv('covid19_utf8.csv')

# Extracting relevant columns and reshaping the data for plotting
columns_of_interest = [col for col in data.columns if '전체' in col]
reshaped_data = pd.DataFrame({
    '지역구': [col.replace(' 전체', '') for col in columns_of_interest],
    '확진자': [data[col].sum() for col in columns_of_interest]
})

# Plotting the barplot
plt.figure(figsize=(10, 6))
sns.barplot(
    x="지역구",
    y="확진자",
    data=reshaped_data,
    order=reshaped_data.sort_values(by="확진자", ascending=False)['지역구'],
    palette="Blues_r",
    hue=None,  # Explicitly setting hue to None
    dodge=False
)

plt.xlabel('지역구')
plt.ylabel('확진자')
plt.title('지역별 확진자 수')
plt.xticks(rotation=45)
plt.legend([], [], frameon=False)  # Removing legend to avoid warnings
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib as mpl
import numpy as np
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# Set font to NanumGothic
mpl.rcParams['font.family'] = 'NanumGothic'
mpl.rcParams['axes.unicode_minus'] = False  # Prevent issues with minus sign

# Load the data
data = pd.read_csv('covid19_utf8.csv')

# Adding month data for demonstration purposes
np.random.seed(42)
data['월'] = np.random.choice([6, 7, 8], size=len(data))

# Extracting relevant columns and reshaping the data for plotting
columns_of_interest = ['종로구 전체', '강남구 전체', '관악구 전체', '용산구 전체', '월']
filtered_data = data[columns_of_interest]
reshaped_data = pd.melt(filtered_data, id_vars=['월'], var_name='지역구', value_name='확진자')
reshaped_data['지역구'] = reshaped_data['지역구'].str.replace(' 전체', '')

# Remove top 10% of outliers
upper_quantile = reshaped_data['확진자'].quantile(0.70)
reshaped_data = reshaped_data[reshaped_data['확진자'] <= upper_quantile]

# Plotting the barplot
plt.figure(figsize=(15, 6))

# Plotting boxplot for each month by region
plt.subplot(1, 2, 1)
sns.boxplot(x='월', y='확진자', hue='지역구', data=reshaped_data)

plt.ylabel('확진자')
plt.title('월별 지역구별 확진자 수')

# Plotting boxplot for each region by month
plt.subplot(1, 2, 2)
sns.boxplot(x='지역구', y='확진자', hue='월', data=reshaped_data)

plt.ylabel('확진자')
plt.title('지역구별 월별 확진자 수')

plt.tight_layout()
plt.show()
```

```python
# 1. 팀별 salary 분포 시각화
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import os

# Load the salary data
data = pd.read_csv('NBA_season1718_salary.csv')

# Remove unnamed index column if it exists
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]

# Plotting salary distribution by team
plt.figure(figsize=(15, 8))
sns.boxplot(x='Tm', y='season17_18', data=data)
plt.xticks(rotation=45)
plt.xlabel('Team')
plt.ylabel('Salary (in USD)')
plt.title('Salary Distribution by Team (2017-2018 Season)')
plt.tight_layout()
plt.show()

# 2. 포지션과 salary 상관 관계 분석
# Load the player data
player_data = pd.read_csv('player_data.csv')

# Merge salary data with player data based on player name
merged_data = pd.merge(data, player_data, left_on='Player', right_on='name', how='inner')

# Extract relevant columns for analysis
merged_data = merged_data[['Player', 'Tm', 'season17_18', 'position']]

# Plotting salary distribution by position
plt.figure(figsize=(10, 6))
sns.boxplot(x='position', y='season17_18', data=merged_data)
plt.xlabel('Position')
plt.ylabel('Salary (in USD)')
plt.title('Salary Distribution by Position (2017-2018 Season)')
plt.tight_layout()
plt.show()

# 3. PER과 salary 상관 관계 분석"
# Load the unzipped CSV file
season_stats = pd.read_csv('Seasons_Stats.csv')

# Fill missing values in relevant columns with the mean
season_stats['PER'] = season_stats['PER'].fillna(season_stats['PER'].mean())
season_stats['Year'] = season_stats['Year'].fillna(season_stats['Year'].mean())

# Merge salary data with season stats data based on player name and year (considering relevant years only)
merged_stats_salary = pd.merge(data, season_stats[season_stats['Year'] >= 2014], on='Player', how='inner')

# Extract relevant columns for analysis
merged_stats_salary = merged_stats_salary[['Player', 'Tm_x', 'season17_18', 'PER']]

# Plotting the relationship between PER and salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PER', y='season17_18', data=merged_stats_salary)
plt.xlabel('Player Efficiency Rating (PER)')
plt.ylabel('Salary (in USD)')
plt.title('Relationship between PER and Salary (2017-2018 Season)')
plt.tight_layout()
plt.show()
```
